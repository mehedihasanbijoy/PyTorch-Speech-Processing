Where,
- **A - Amplitude:** The amplitude (A) is the maximum displacement of the waveform from its equilibrium position (the equilibrium position refers to the neutral state of a speech signal, where the vocal tract and associated components are at rest, and there is no significant change or articulatory movement). It controls the "height" of the oscillation. A larger amplitude results in a taller wave, while a smaller amplitude produces a shorter wave. It affects the intensity or strength of the oscillation.
- **sin - Sine Function:** The sine function is a trigonometric function that produces a smooth, periodic oscillation between -1 and 1. The sine function gives the waveform its characteristic shape. It determines how the amplitude varies over time, creating a smooth oscillatory pattern.
- **2πf - Angular Frequency:** Angular frequency (2πf) is the rate at which the waveform oscillates in radians per unit of time. It controls the frequency of the wave. A higher angular frequency (2πf) results in a higher number of oscillations per unit of time (frequency f), affecting how quickly the wave completes cycles.
- **t - Time:** Time (t) is the independent variable representing the moment at which you are evaluating the waveform. As time changes, the sine wave oscillates according to the specified frequency and amplitude. It determines the position of the wave along the time axis.
- **φ - Phase Shift:** Phase shift (φ) represents a horizontal translation of the waveform along the time axis. It determines the starting point of the wave at t=0. A positive phase shift (φ) means a shift to the right, and a negative phase shift means a shift to the left. It affects the horizontal position of the waveform.

**Let's walk through a specific example using the sine wave equation:**

y(t) = 3 sin⁡(2π⋅2t+π/4)

In this example:

- A=3: The amplitude is set to 3, meaning the wave will oscillate between -3 and 3.
- f=2: The frequency is set to 2 Hz, indicating that the wave completes 2 cycles per second.
- φ=π/4​: The phase shift is π/4​, meaning the wave starts at a phase angle of π/4​.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
A = 3
f = 2
lambda_value = np.pi / 4  # Phase shift in radians

# Generate time values from 0 to 2 seconds with small time steps
t = np.arange(0, 2, 0.01)

# Calculate the waveform using the sine wave equation
y = A * np.sin(2 * np.pi * f * t + lambda_value)

# Plot the waveform
plt.figure(figsize=(8, 4))
plt.plot(t, y, label=r'$3 \sin(2 \pi \cdot 2t + \frac{\pi}{4})$')
plt.title('Sine Wave Example')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```

# Frequency

- Frequency is a fundamental property of a waveform and refers to the number of cycles the waveform completes in one second, measured in Hertz (Hz).
- In the context of sound, frequency is closely associated with the perceptual quality of pitch (highness or lowness of voice (nothing related to loudness/amplitude)). Higher frequencies are typically perceived as higher pitches, while lower frequencies are associated with lower pitches.

# Amplitude

- Amplitude in the context of a waveform represents the strength or intensity of the signal. It is a measure of the maximum displacement of particles in a medium (such as air for sound waves) from their equilibrium positions during one complete cycle of the wave.
- Amplitude is often measured in decibels (dB) in the context of sound waves.
- Amplitude is directly related to the perceived loudness of a sound. Higher amplitudes are associated with louder sounds, while lower amplitudes are perceived as softer or quieter.

# Phase

- Phase represents the temporal position of the waveform within its cycle. It is measured in radians and provides information about where the waveform is in its oscillation at a specific point in time.
- The phase is closely related to its frequency. Changes in phase correspond to changes in the frequency of the signal.
- Speech signals can be analyzed using a spectrogram, which visualizes the frequency content over time.
    - The phase spectrogram complements the traditional spectrogram by representing the phase information at different frequencies over time. This can be valuable for detailed analysis of speech characteristics.

# Frequency Vs Amplitude

- Frequency and amplitude are fundamentally different in their definitions and units. Frequency is the number of cycles per unit of time (measured in Hertz), while amplitude is the maximum displacement of a wave from its equilibrium position.

- While both frequency and amplitude contribute to the perceptual characteristics of waves, they affect different aspects of perception. Frequency influences the perceived pitch, while amplitude influences the perceived loudness.

- In mathematical representations of waves, frequency is associated with the rate of oscillation or the number of cycles per unit of time. Amplitude, on the other hand, is associated with the maximum value of the wave's displacement from its equilibrium position.

- Frequency corresponds to how rapidly a wave oscillates, representing the temporal aspect of the wave. Amplitude, on the other hand, represents the strength or intensity of the wave, reflecting its energy content.

# An Example to Demonstrate Frequency Vs Amplitude

If you speak the same word faster and slower, the frequency of the corresponding sound waves will change. Consider the following examples:

- **H..E..L..L..O** (Slower)
- **HELLO** (Faster)

**Changes in Frequency:**
- **Faster Speech (Higher Speed):** Speaking the word faster increases the frequency of the sound waves. Frequency is the number of cycles per unit of time, and faster speech results in more cycles occurring in a given time, leading to a higher frequency.

- **Slower Speech (Lower Speed):** Conversely, speaking the word more slowly decreases the frequency of the sound waves. Slower speech allows fewer cycles to occur in the same amount of time, resulting in a lower frequency.

- **Softer Speech:** When saying "hello" softer, the amplitude of the sound waves decreases. The vocal folds vibrate with less intensity, leading to smaller displacements of air particles. The peaks and troughs in the waveform become less pronounced, indicating a smaller amplitude.


# Frequency, Amplitude, and Phase

 - Definition and Measurement   
    - **Frequency:** Represents the number of cycles per unit of time (measured in Hertz).
    - **Amplitude:** Represents the maximum displacement from equilibrium (measured in dB).
    - **Phase:** Represents the position within a cycle (measured in radians or degrees)

 - Perceptual Characteristics   
    - **Frequency:** Affects the perceived pitch of a sound.
    - **Amplitude:** Affects the perceived loudness or intensity of a sound.
    - **Phase:** Primarily influences the timing or synchronization of waveforms.

 - Mathematical Representation   
    - **Frequency:** Typically represented as a scalar value (e.g., 440 Hz).
    - **Amplitude:** Represented as the height of the wave (peak or trough).
    - **Phase:** Represented as an angular measurement (radians or degrees).

 - Manipulation in Signal Processing   
    - **Frequency:** Manipulated in tasks like filtering, pitch-shifting, and modulation.
    - **Amplitude:** Manipulated in tasks like volume control, dynamic range compression, and amplification.
    - **Phase:** Manipulated in tasks like phase shifting, time-domain modifications, and synchronization.

# Pitch

- Pitch is a perceptual attribute of sound that corresponds to the perceived frequency of a sound wave.

- Higher pitch is associated with higher frequencies, and lower pitch is associated with lower frequencies.

- While frequency is the objective measurement of the number of cycles per second (Hertz), pitch is the subjective perception of that frequency.

- We don't hear frequency in a linear way, rather in a logarithmic way. Two frequencies are perceived similarly if they differ by a power of 2.
